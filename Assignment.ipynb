#Pandas Problems

#Q1

```python
#Basic Exploration
display(data.head(10))
print("Shape (rows, columns):", data.shape)
display(data.describe())
```

#Q2

```python
#Filtering and Indexing
filtered_data = data[(data["model year"] == 75) & (data["weight"] < 3000)][["car name", "weight", "mpg"]]
display(filtered_data)
```

#Q3

```python
#Handling Missing Data
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
data["horsepower"] = pd.to_numeric(data["horsepower"], errors="coerce")
median_hp = data["horsepower"].median()
data["horsepower"] = data["horsepower"].fillna(median_hp)
print("Median horsepower:", median_hp)
```

#Q4

```python
#Data Transformation
data["p2w_ratio"] = data["horsepower"] / data["weight"]
display(data[["horsepower", "weight", "p2w_ratio"]].head())
```

#Q5

```python
#Group By
avg_mpg_by_origin = data.groupby("origin")["mpg"].mean()
display(avg_mpg_by_origin)
```

#Q6

```python
#Sorting
def performance_score(row):
    return row["mpg"] * row["acceleration"] / row["weight"]


t10_mpg = data.sort_values(by="mpg", ascending=False).head(10)
display(t10_mpg[["car name", "mpg", "acceleration", "weight"]])
```

#Q7

```python
#Applying Function
data["performance_score"] = data.apply(performance_score, axis=1)
display(data[["car name", "mpg", "acceleration", "weight", "performance_score"]].head())
```

#Q8

```python
#Visualization Preparation
summary_year = (data.groupby("model year")[["mpg", "weight", "horsepower"]].mean().reset_index())
summary_year.rename(columns={"mpg": "avg_mpg", "weight": "avg_weight", "horsepower": "avg_horsepower"},inplace=True,)
display(summary_year)
```

#Q10

```python
#Finding Anomalies
Q1=np.percentile(data['mpg'],25)
Q2=np.percentile(data['mpg'],50)
Q3=np.percentile(data['mpg'],75)
IQR=Q3-Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data["mpg"] < lower_bound) | (data["mpg"] > upper_bound)][["car name", "mpg", "model year"]]
display(outliers)
```

#Matplotlib Problems

#Q1

```python
import matplotlib.pyplot as plt
```

```python
#Distribution of miles per gallon
plt.hist(data["mpg"], bins=20, edgecolor="black")
plt.xlabel("Miles per Gallon (mpg)")
plt.ylabel("Frequency")
plt.title("Distribution of MPG in the Dataset")
plt.show()
```

#Q2

```python
#mpg vary with the number of cylinders
data.boxplot(column="mpg", by="cylinders", grid=False)
plt.xlabel("Number of Cylinders")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("MPG by Number of Cylinders")
plt.suptitle("")
plt.show()
```

#Q3

```python
#relationship between horsepower and mpg
data["horsepower"] = pd.to_numeric(data["horsepower"], errors="coerce")
data_hp = data.dropna(subset=["horsepower"])
plt.scatter(data_hp["horsepower"], data_hp["mpg"], alpha=0.7)
plt.xlabel("Horsepower")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("Horsepower vs. MPG")
plt.show()
```

#Q4

```python
#car weight influence mpg
plt.scatter(data["weight"], data["mpg"], alpha=0.7, label="Data points")
z = np.polyfit(data["weight"], data["mpg"], 1)
p = np.poly1d(z)
plt.plot(data["weight"], p(data["weight"]), color="red", linewidth=2, label="Trend line")
plt.xlabel("Weight")
plt.ylabel("Miles per Gallon (mpg)")
plt.title("Weight vs. MPG with Trend Line")
plt.legend()
plt.show()
```

#Q5

```python
#average mpg across model years
avg_mpg_per_year = data.groupby("model year")["mpg"].mean()
plt.plot(avg_mpg_per_year.index, avg_mpg_per_year.values, marker="o")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.title("Average MPG Across Model Years")
plt.grid(True)
plt.show()
```

#Q6

```python
#count of cars distributed by origin
origin_counts = data["origin"].value_counts().sort_index()
origin_labels = {1: "USA", 2: "Europe", 3: "Japan"}
plt.bar(origin_counts.index.map(origin_labels), origin_counts.values)
plt.xlabel("Origin")
plt.ylabel("Number of Cars")
plt.title("Number of Cars by Origin")
plt.show()
```

#Q7

```python
#acceleration values vary across different cylinders
data.boxplot(column="acceleration", by="cylinders", grid=False)
plt.xlabel("Number of Cylinders")
plt.ylabel("Acceleration")
plt.title("Acceleration by Number of Cylinders")
plt.suptitle("")
plt.show()
```

#Q8

```python
#year with most number of car entries
car_counts_by_year = data["model year"].value_counts().sort_index()
plt.bar(car_counts_by_year.index, car_counts_by_year.values)
plt.xlabel("Model Year")
plt.ylabel("Number of Cars")
plt.title("Number of Cars by Model Year")
plt.show()

most_cars_year = car_counts_by_year.idxmax()
most_cars_count = car_counts_by_year.max()
print(f"The year with the most car entries is {most_cars_year} with {most_cars_count} cars.")
```

```python
#clustering pattern among weight, horsepower, and mpg
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(
    data_hp["weight"],
    data_hp["horsepower"],
    data_hp["mpg"],
    c=data_hp["mpg"],
    cmap="viridis",
    alpha=0.7,)
ax.set_xlabel("Weight")
ax.set_ylabel("Horsepower")
ax.set_zlabel("MPG")
ax.set_title("3D Scatter Plot: Weight vs Horsepower vs MPG")
plt.show()
```
